## 4. Conditions: HOT or NORMAL days
### ðŸ”¶ Task 4.1 â€“ Classify SST days

```python
threshold = 28.5
sst = [25, 26, 27, 26, 23, 29, 26]
for i in range(len(sst)):
    temp = sst[i]
    day = i + 1
    # TODO: write if condition here
    if temp > threshold:
        print("Day", day, ": HOT")
    if temp < threshold:
        print("Day", day," : Normal")
```

## 5. Filtering SST values
### ðŸ”¶ Task 5.1 â€“ Create HOT and NORMAL SST lists

```python
sst = [25, 26, 27, 26, 23, 29, 26]
hot_sst = []
normal_sst = []
threshold = 27
for i in range(len(sst)):
    temp = sst[i]
    if temp > threshold:
        hot_sst.append(temp)
    if temp < threshold:
        normal_sst.append(temp)

print("Hot SST values:", hot_sst)
print("Normal SST values:", normal_sst)
```

## 6. Mini Task â€“ Rainfall classification
### ðŸ”¶ Task 6.1 â€“ Classify rainfall intensity

```python
rain = [0.0, 12.5, 34.0, 2.0, 0.0, 55.1, 8.4]

no_rain = 0
light = 0
moderate = 0
heavy = 0

for r in rain:
    if r == 0:
        no_rain += 1
    elif r > 0 and r <= 10:
        light += 1
    elif r > 10 and r <= 50:
        moderate += 1
    else:
        heavy += 1

print("No rain days:", no_rain)
print("Light rain days:", light)
print("Moderate rain days:", moderate)
print("Heavy rain days:", heavy)
```
