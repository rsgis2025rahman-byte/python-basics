# üå° What are Anomalies?

In climate and oceanography, an anomaly is:

\[ \, 	ext{anomaly} = 	ext{value} - 	ext{mean} \, \]

Used for:
- marine heatwave detection  
- climate trend analysis  
- seasonal cycle removal  
- comparing conditions to long‚Äëterm averages  

---

## üî∂ Task 3.1 ‚Äî Compute SST Anomalies

Use:

```python
sst = np.array([28.1, 28.3, 28.4, 28.0, 27.9, 28.5, 28.7, 28.6, 28.2, 28.1])
```

Steps:
1. Compute mean  
2. Compute anomalies using vectorized subtraction  
3. Print both  


````python
import numpy as np
sst = np.array([28.1, 28.3, 28.4, 28.0, 27.9, 28.5, 28.7, 28.6, 28.2, 28.1])

mean_sst = np.mean(sst)
sst_anom = sst - mean_sst

print("Mean SST:", mean_sst)
print("SST Anomalies:", sst_anom)
````

# 4. Plotting Scientific Time Series

Scientific plots help us:
- observe trends  
- detect extreme events  
- compare variables  
- spot errors or anomalies  
- understand system behavior  

A **time series plot** shows how a variable changes with time.

---

## Elements of a Good Plot
- **x‚Äëaxis**: time (days, months, years)  
- **y‚Äëaxis**: scientific variable  
- **markers**: to highlight points  
- **grid**: improves readability  
- **title**: explains what you're seeing  
- **labels**: indicate units  

---

## üî∂ Task 4.1 ‚Äî Plot the SST Time Series

Steps:
1. Create day numbers:  
   ```python
   days = np.arange(1, 11)
   ```  
2. Plot using markers  
3. Add title, axis labels, grid  


````python
import numpy as np
import matplotlib.pyplot as plt
days = np.arange(1, 11)
sst = [28.1, 28.3, 28.4, 28.0, 27.9, 28.5, 28.7, 28.6, 28.2, 28.1]

plt.figure()
plt.plot(days, sst, marker='*')
plt.xlabel("Day")
plt.ylabel("SST (¬∞C)")
plt.title("Daily SST Time Series")
plt.grid(True)
plt.show()
````

# 5. Plotting Anomalies
